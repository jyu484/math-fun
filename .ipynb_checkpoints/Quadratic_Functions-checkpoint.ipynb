Quadratic Functions

```python
#I need these modules for the functions to work.
import numpy as np
import cmath
```

```python
#Type the coefficients here. a, b, and c correspond to the quadratic function in standard form.
a = 2
b = -8
c = -10
```

```python
#I am creating a function to calculate the location of the vertex.
def quadratic_vertex(a, b, c):
    '''This function will determine the vertex of a quadratic function using three inputs.
    The inputs are the coefficients in standard form f(x)=ax^2+bx+c.
    Written by John Yu
    '''
    
    #I want to make sure all the inputs are floats or intgers.
    if isinstance(a, (int, float)) and isinstance(b, (int, float)) and isinstance(c,(int, float)):
        pass

        #This section will calculate both the x- and y-values of the vertex.
        vertex_x = -b/(2*a)
        vertex_y = a*(vertex_x)**2+b*(vertex_x)+c
        
        return vertex_x, vertex_y
    
    else:
        raise(TypeError("The input arguments must be integers or floats."))

help(quadratic_vertex)
```

```python
#I am creating a function to calculate the roots.
def quadratic_roots(a, b, c):
    '''This function will determine the roots of a quadratic function using three inputs.
    The inputs are the coefficients in standard form f(x)=ax^2+bx+c.
    Written by John Yu
    '''

    #I want to make sure all the inputs are floats or intgers.
    if isinstance(a, (int, float)) and isinstance(b, (int, float)) and isinstance(c,(int, float)):
        pass

        #The rest of the code will rely on the quadratic formula.
        #The discriminant determines the conditions of the code.
        discriminant = b**2-4*a*c

        #There are two real solutions when the discriminant is greater than 0.
        if discriminant > 0:
            root1 = (-b + np.sqrt(discriminant))/(2*a)
            root2 = (-b - np.sqrt(discriminant))/(2*a)
            
            return root1, root2

        #There is one real solution when the discrimant is 0.
        if discriminant == 0:
            root1 = (-b+np.sqrt(discriminant))/(2*a)
            
            return root1, root1

        #There are two complex solutions when the discriminant is less than 0.
        if discriminant < 0:
            root1 = (-b + cmath.sqrt(discriminant))/(2*a)
            root2 = (-b - cmath.sqrt(discriminant))/(2*a)
            
            return root1, root2

    else:
        raise(TypeError("The input arguments must be integers or floats."))

help(quadratic_roots)
```

```python
#I am going to print the vertex outside of the function's code.
vertex = quadratic_vertex(a, b, c)
print(f'The vertex is located at ({vertex[0]}, {vertex[1]}).')
```

```python
#I am going to print the roots outside of the function's code.
roots = quadratic_roots(a, b, c)
print(f'The roots are located at ({roots[0]}, 0) and ({roots[1]}, 0).')
```
