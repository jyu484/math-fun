Fourth-Order Runge-Kutta Method

```python
#I need these modules.
import numpy as np
```

```python
#These are the initial conditions of the problem. Aside from the equation, the other values need to be integers or floats.
y_prime = lambda x, y: x - y
x_0 = 0
y_0 = 1
h = 0.1
```

```python
def fourth_order(y_prime, x_0, y_0, h, x_n):
    '''
    This function will perform the Fourth-Order Runge-Kutta Method given five inputs.
    The first input arugment is the differential equation, y_prime.
    The second input is the initial x-value, x_0.
    The third input is the initial y-value, y_0.
    The fourth input is the step-size, h.
    The fifth input is the final x-value, x_n.
    Written by John Yu
    '''

    #The lines of code will make sure that the inputs are integers or floats.
    if isinstance(x_0, (int, float)) and isinstance(y_0, (int, float)) and isinstance(h, (int, float)) and isinstance(x_n, (int, float)):
        pass
        
        #These lines of code will generate the arrays that will store the x-values and y-values.
        x_values = np.arange(x_0, x_n + h, h)
        y_values = np.zeros(len(x_values))
        y_values[0] = y_0
    
        #These lines of code will calculate the y-values using the formulas from the Fourth-Order Runge-Kutta Method and add them to the array.
        for i in range(0, len(x_values) - 1):
            k_1 = y_prime(x_values[i], y_values[i])
            k_2 = y_prime(x_values[i] + (h/2), y_values[i] + (1/2) * k_1 * h)
            k_3 = y_prime(x_values[i] + (h/2), y_values[i] + (1/2) * k_2 * h)
            k_4 = y_prime(x_values[i] + h, y_values[i] + k_3 * h)
            y_values[i + 1] = y_values[i] + (h/6) * (k_1 + 2*k_2 + 2*k_3 + k_4)

        return x_values, y_values

    else:
        raise(TypeError("The initial values need to be integers or floats."))
              
#This will explain how to use the function.
help(fourth_order)
```

```python
fourth_order(y_prime, x_0, y_0, h, 0.5)
```
